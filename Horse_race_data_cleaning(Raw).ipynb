```python
import numpy as np
def avg_displacement(df,latlong):
  difference=np.diff(df[latlong])
  difference=difference.tolist()
  difference=[abs(number) for number in difference]
  difference.insert(len(difference),difference[len(difference)-1])
  
  return difference
```

```python
import numpy as np
df_merged1['latdist_covered']=avg_displacement(df_merged1,'latitude')
df_merged1['latdist_covered']= df_merged1['latdist_covered']*111139
df_merged1['longdist_covered']=avg_displacement(df_merged1,'longitude')
df_merged1['longdist_covered']= df_merged1['longdist_covered']*111139
df_merged1['displacement']= np.sqrt(df_merged1['latdist_covered']**2+df_merged1['longdist_covered']**2)
df_merged1['m/s']=df_merged1['displacement']
```

```python
conditions = [
    (Grouped['dense_rank']==1),
    (Grouped['dense_rank']!=1)
]

values = [1,0]

Grouped['Winner'] = np.select(conditions, values)
```
